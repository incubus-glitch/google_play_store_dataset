## Задачи для классификации:
- Предсказать, является ли приложение Free / Paid (target: `Type`)
- Предсказать категорию приложения Game, Education, Tools и т.д. (target: `Category`)
- Предсказать уровень контента (target: `Content Rating`)

```python
df.shape
```

```python
X = df.drop('Type', axis=1)
X['Last Updated Year'] = X['Last Updated'].dt.year
X['Last Updated Month'] = X['Last Updated'].dt.month
X['Last Updated day'] = X['Last Updated'].dt.day
X = X.drop('Last Updated', axis=1)
X = X.drop('Price', axis=1)
```

```python
le = LabelEncoder()
y = le.fit_transform(df['Type'])
print(le.classes_)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=.2,
    random_state=42, 
    stratify=y
)
```

```python
num_cols = X_train.select_dtypes(include=['int64', 'float64']).columns
cat_cols = X_train.select_dtypes(include=['string']).columns
```

```python
preprocessor_num = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore'), cat_cols)
    ]
)

preprocessor_tree = ColumnTransformer(
    transformers=[
        ('num', 'passthrough', num_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore'), cat_cols)
    ]
)
```

```python
pipe = make_pipeline(
    preprocessor_tree,
    DecisionTreeClassifier(random_state=42)
)
```

```python
pipe.get_params().keys()
```

```python
param_grid = {
    'decisiontreeclassifier__criterion': [
        'gini',
        'entropy',
        'log_loss'
    ],

    'decisiontreeclassifier__max_depth': [
        3, 5, 10, 15, 20, None
    ],

    'decisiontreeclassifier__min_samples_split': [
        2, 5, 10, 20
    ],

    'decisiontreeclassifier__min_samples_leaf': [
        1, 2, 5, 10
    ]
}
```

```python
grid = GridSearchCV(
    estimator=pipe,
    param_grid=param_grid,
    scoring='f1',
    cv=5,
    n_jobs=-1,
    verbose=2
)
```

```python
grid.fit(X_train, y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```

```python
y_pred = grid.predict(X_test)
print(classification_report(y_test, y_pred))
```
